```python
import pandas as pd

filename = 'nyc_subway_weather.csv'
subway_df = pd.read_csv(filename)
def correlation(x, y):
    '''
    Fill in this function to compute the correlation between the two
    input variables. Each input is either a NumPy array or a Pandas
    Series.
    
    correlation = average of (x in standard units) times (y in standard units)
    
    Remember to pass the argument "ddof=0" to the Pandas std() function!
    '''
    std_x = (x - x.mean()) / x.std(ddof=0)
    std_y = (y - y.mean()) / y.std(ddof=0)
    
    # * does vectorized multiplication
    return (std_x * std_y).mean()
```

```python
def convert_grades(grades):
    '''
    Fill in this function to convert the given DataFrame of numerical
    grades to letter grades. Return a new DataFrame with the converted
    grade.
    
    The conversion rule is:
        90-100 -> A
        80-89  -> B
        70-79  -> C
        60-69  -> D
        0-59   -> F
    '''
    if grades >= 90:
        return 'A'
    elif grades >= 80:
        return 'B'
    elif grades >= 70:
        return 'C'
    elif grades >= 60:
        return 'D'
    else:
        return 'F'
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline

filename = 'nyc_subway_weather.csv'
subway_df = pd.read_csv(filename)

grouped_data = subway_df.groupby(['latitude', 'longitude'], as_index=False).mean()
bubble = (grouped_data.ENTRIESn)/grouped_data.ENTRIESn.std(ddof=0)
plt.scatter(grouped_data['longitude'], grouped_data['latitude'], s=bubble*15, c=grouped_data.ENTRIESn, cmap='bwr')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('NY Subway Scatterplot: Bubbble size depends on the number of hourly entries')
```

```python
# difference between cut and qcut ?
grouped_by_tempi = subway_df.groupby(pd.cut(\
    subway_df['meantempi'],\
    bins = [45, 50, 55, 60, 65, 70, 75, 80],\
    labels = ['50F', '55F', '60F', '65F', '70F', '75F', '80F']
))

means =  grouped_by_tempi.mean()[['ENTRIESn_hourly', 'EXITSn_hourly']]
plot = means.plot(kind = 'bar', title = 'Entries and Exits hourly per temperature')
plot.set_xlabel('Mean temperature')
plot.legend(['Hourly entries', 'Hourly exits'])
```
